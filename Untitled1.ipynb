```python
import matplotlib.pyplot as plt
seq = []
aes = [-1.1,-1, -0.5, 0, 0.5, 0.7, 1, 1.1]
for i in range(len(aes)):
    x = 1
    a = aes[i]
    seq.append([])
    for r in range(20):
        seq[i].append(x)
        x = x *a
    
for s  in seq:
    plt.plot(s)
plt.show()
```

```python
# Ejercicio 11

def new_x(x):
    return  2 * x - x** 2
seeds = [ 0.5, 0, 0.5, 1, 1.5]
seqs = []
for i, seed in zip(range(len(seeds)), seeds):
    seqs.append([seed])
    for r in range(5):
        x = new_x(seqs[i][-1])
        seqs[i].append(x)
print (seqs)
for seq in seqs:
    plt.plot(seq)
plt.show()
```
